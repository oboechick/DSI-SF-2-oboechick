```python
def plot_heat_map(cnt_plus_total):
    latitude = cnt_plus_total.latitude.values
    longitude = cnt_plus_total.longitude.values
    total_number_of_resp = cnt_plus_total.total_number_of_resp.values
    
    world_map = folium.Map(location=[49.8175, 18.4730], zoom_start=3)
    data = [(x,y,d) for x,y,d in zip(latitude,longitude,total_number_of_resp)]
    world_map.add_children(plugins.HeatMap(data,radius=12))
    return world_map
```

```python
plot_heat_map(cnt_plus_total)
```
